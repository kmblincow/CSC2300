#**filter(): Keep rows matching specific criteria**  
The filter() function in the dplyr package allows us to subset our data frame so that we only keep rows based on certain conditions.  
  
As a reminder from our early days doing the Lakers assignment, if we want to look for specific values we write the column name "==" the values we want. When filtering we use "==" not "=".
  
Let's begin by demonstrating how we can filter for rows associated with TV shows that were sourced from Manga.

```r
#use filter() to keep rows associated with TV shows that were sourced from Manga.
anime %>% filter(type == "TV" & source == "Manga")
```

Notice how the only values in the type column are "TV" and the source column are "Manga"? We kept only rows where those conditions were true.  
NOTE: we can use comma or ampersand (&) to represent the AND condition.  
  
We can use the "|" symbol for the OR condition.  
For example, if we want to filter for rows where genre was "Adventure" **or** "Drama" we can do so using the code below.

```r
#use filter() to keep rows where the genre is Adventure OR Drama
anime %>% filter(genre == "Adventure" | genre == "Drama")
```

We can also combine conditions in a single filter call. For example, if we wanted to keep rows from TV shows sourced from Manga that were in the Adventure or Drama genres we can run the code chunk below.

```r
#filter for Managa sourced Adventure or Drama TV shows
anime %>%
  filter(type == "TV" & source == "Manga" & (genre == "Adventure" | genre == "Drama"))
```

#**select(): Pick columns by name**  
  
The select() function allows us to select specific columns. We can select columns by writing out their names directly, but there are also lots of helpful functions and shorthand operators that speed this process up.  
  
We can use the colon (:) to select multiple contiguous columns, as you can see in the code chunk below.

```r
#select all columns from title_english to genre
anime %>% select(title_english:genre)
```

We can also link column selections using commas. For example, we can select the name column and all columns between rank and favorites using the code below.

```r
#select name and all columns between rank and favorites
anime %>% select(name, rank:favorites)
```

There are also a number of helper functions that work with select that allow us to find columns that have names meeting certain conditions. These include: contains(), starts_with(), ends_with(), and matches(). For example, we can select all columns that contain the word "date" using the code chunk below.

```r
#select all columns with the word "date"
anime %>% select(contains("date"))
```

We can also remove columns using the "-" symbol. This is useful is you want to keep most of your columns and only drop one or two of them. For example, the "X" column is an artifact of loading the csv file. We can get rid of it using the code chunk below.

```r
#remove the X column
anime %>% select(-X)
```

#**"Chaining" or "Piping"**  
  
Typically if you wanted to perform mutiple operations in one go in R you would need to nest functions within each other. This is confusing and difficult to read.  
  
The pipe operators (%>% or |>) allow you to write commands in a natural order that is more intuitive and easier to read. You can imagine the pipe operator to mean "and then".  
  
For example, say we wanted to select the name and rank columns, then filter so we only have titles that are ranked in the top 10, and then only show rows with distinct values for those column. We can accomplish that using the nesting syntax using the code chunk below.

```r
#accomplish multiple tasks at once using nesting
distinct(filter(select(anime, name, rank), rank < 11))
```

That's all well and good, but it's hard to force your brain to work from the inside of that code out following the nesting structure.  
A more intuitive way to accomplish the same task is to use pipe operators to sequentially accomplish the same tasks as shown in the code chunk below.

```r
#accomplish multiple tasks at once using piping
anime %>% #"take our anime data and then..."
  select(name, rank) %>% #"select the name and rank columns and then..."
  filter(rank < 11) %>% #"filter so we only have the top 10 titles and then..."
  distinct() #"only show rows with distinct combinations values from all columns"
```

You can hopefully see how piping increases readability significantly when there are many commands.

#**mutate(): Add new variables**  
The mutate() function allows us to add new variables to a data frame. These variables can be standalone, or they can rely on existing variables.  
  
As an example what I mean by a "standalone" new variable, you can see in the code chunk below that I am creating a new column called "one" that is populated with the value 1.

```r
#add a column called "one" that is filled with 1s
anime %>% mutate(one = 1)
```

But what is cooler, is that we can use values from other columns to perform calculations for our new column. For example, let's say we want to scale the popularity metric to be from 0 to 1, because it's hard to wrap our head around what 15468 popularity means (FYI that's the highest popularity score). We can do this by dividing all the popularity scores by the max popularity score, as you can see in the code chunk below.

```r
#create a new column that is the scaled popularity score
anime %>% mutate(popularity_scaled = popularity/max(popularity))
```

Note that all the code we have been running thus far is still just printing our manipulations and not saving them. If we want to store our new column, we need to use the assignment arrow, as seen in the code chunk below.

```r
#overwrite our original data with the data that includes the new popularity_scaled column
anime <- anime %>% mutate(popularity_scaled = popularity/max(popularity))
```

#**group_by() & summarize(): Reduce variables to summary values**  
Many times we want to get summaries of our data. We can do that using summarize() (or summarise(), both work). We can use summarize() on it's own to get summary values for specific columns, as shown in the code chunk below where we calculate the mean number of episodes of anime.

```r
#calculate mean number of episodes
anime %>% summarize(mean_episodes = mean(episodes, na.rm = T))
```

NOTE: We need to add the "na.rm = T" argument to the mean function, because there are NAs in this column and R doesn't know how to handle that. We will learn more about NAs next week.  
  
Importantly, you see that summarize() completely changes the data frame. It collapses it into a summary of a single column. We can also link multiple summaries using commas in the summarize() function. We will demonstrate this in the code chunk below where we calculate the mean, min, and max number of episodes.

```r
#calculate mean, min, and max number of episodes
anime %>% summarize(mean_episodes = mean(episodes, na.rm = T),
                    min_episodes = min(episodes, na.rm = T),
                    max_episodes = max(episodes, na.rm = T))
```

You see that now our summary has three columns, one for each of the summary statistics we calculated (mean, min, and max).  
  
    
Often summarize() is used in conjunction with group_by(). group_by() creates an implicit grouping in your data frame based on a categorical variable. When summarize() comes after group_by(), the summary will be calculated at each level of the grouping variable.  
  
For example, in the code chunk below we will calculate the mean, min, and max number of episodes for each studio putting out anime.

```r
#calculate mean, min, and max episodes by genre
anime %>%
  group_by(studio) %>%
  summarize(mean_episodes = mean(episodes, na.rm = T),
            min_episodes = min(episodes, na.rm = T),
            max_episodes = max(episodes, na.rm = T))
```

Now you can see we have a row for each studio and columns showing the summary values for each studio.  
  
There are also helper functions for summarize(). These include n() which counts the number of rows in each group and n_distinct() which counts the number of unique items in a given vector.  
We might use these to count the number of anime made by each studio, and the number of distinct producers for each studio, as shown in the code chunk below.

```r
#print the number of anime made by each studio and number of distinct producers for each studio
anime %>%
  group_by(studio) %>%
  summarize(n_anime = n(),
            n_producers = n_distinct(producers))
```
