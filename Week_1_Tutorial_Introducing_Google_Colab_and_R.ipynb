##**What is R?**
R is an open source programming language that is optimized for conducting statistical analyses and making visualizations. It is one of the two most commonly used languages in Data Science, the other being Python. In this course, you will not have to download R or RStudio, which are software tools used to implement the R coding language. ##**Important R Basics**

*   A **function** is a command that tells R to complete a task. For example, the function sample() tells R to draw a random sample.
*   **Base R** is what we collectively call the default functions that exist in R.
*   A **package** is a collection of related functions (and sometimes data) that you can load to extend the capabilities of Base R.
*   **Tidyverse** is a collection of packages that we will be using throughout this course. They share a common philosophy of data and R programming and are designed to work together.
*   We can add **comments** to our code using "#". Anything that comes afer the "#" symbol will not be run as code. This allows us to write normal text along with our code, and is essential for recording explanations of your process and why you are doing what you are doing with your code. I will use these frequently to provide explanations to you. A good rule of thumb is "comment early and often."
